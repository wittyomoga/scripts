### Find which price range of products has max sale. Proce Range below 100, 101-500, 501-1000, above 1000 

```python
# create a new column and populate with <100,101-500,501-1000,>1000 while conditionally checking order price
def get_price_range(price):
    if price < 100:
        return '<100'
    elif  price >= 101 and price <=500:
        return '101-500'
    elif  price >= 501 and price <=1000:
        return '501-1000'
    elif  price > 1000:
        return '>1000'

df['Price Range']=df['Price Each'].apply(lambda x: get_price_range(x))
df.head()
```

```python
df.dtypes
```

```python
# now group by price range to find max quantity ordered
df_qty_prange=df.groupby('Price Range').sum()
df_qty_prange
```
