```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('sales.csv')
df.head()
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
df.head()
```

```python
#sns.distplot(df['Hour'])
sns.distplot(df['Hour'], kde=False) #remove kde 
```

```python
sns.jointplot(x='Sale',y='Hour',data=df, kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='State')
```

```python
sns.rugplot(df['Sale'])
```

```python
sns.kdeplot(df['Sale'])
```

```python
df['City']=df['City']+'-'+df['State']
df_city_sale=df.groupby('City').sum()
df_city_sale
```

```python
sns.kdeplot(df_city_sale['Sale'])
```

```python
sns.barplot(x='City',y='Sale',data=df)
```

```python
sns.boxplot(x='City',y='Sale',data=df)
```
